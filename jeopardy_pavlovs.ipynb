```
import numpy as np

def test_pavlovs(data, pavlov, keywords):
    contains = [data.answer.str.contains(i, case = False) for i in keywords]
    df_keyword = data[np.all(contains, axis =0)]

    # drop first name if pavlov is more than 1 word
    if len(pavlov.split()) > 1:
        pavlov = ' '.join(pavlov.split()[1:])

    df_pavlov = df_keyword[df_keyword.question.str.contains(pavlov,case = False)] 

    if len(df_keyword) == 0:
        score = 0
    else:
        score = len(df_pavlov)/len(df_keyword)
    
    return score, len(df_keyword)
    
scores = {}
for pavlov,clue in pavlovs.items():
    score, length = test_pavlovs(data, pavlov, clue)
    scores[pavlov] = [score, length]
```

```
df_scores = pd.DataFrame(scores, index = ['Score', 'len(df_filtered)']).T
df_scores['keywords'] = [pavlovs[i] for i in list(df_scores.index)]
import plotly.express as px
import pandas as pd
import os

# scores = scores.reset_index()
fig = px.scatter(df_scores,
            x = 'Score', 
            y = 'len(df_filtered)',
            # size = 10,
            hover_data=[df_scores.index, 'keywords'],
            )
fig.update_traces(marker={'size': 15, 'opacity': 0.5})
fig
```

```
fifty_percent = df_scores[df_scores.Score >= 0.5]
fig = px.scatter(fifty_percent,
            x = 'Score', 
            y = 'len(df_filtered)',
            # size = 10,
            hover_data=[fifty_percent.index, 'keywords'],
            )
fig.update_traces(marker={'size': 15, 'opacity': 0.5})
fig
```
